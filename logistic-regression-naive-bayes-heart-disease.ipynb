# **Overview of Dataset & preprocessing**

```python
df = pd.read_csv('/kaggle/input/heart-disease/heart_statlog_cleveland_hungary_final.csv')
df.describe()
```

```python
# getting colums names except target
X = df.drop('target', axis = 1).copy()
X.columns
```

```python
#reshaping from array into a vector
y = df['target'].values.reshape(-1,1)
y.shape
```

```python
#train test split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 43)
```

```python
#normalizing the features
X_train_scaled = StandardScaler().fit_transform(X_train)
X_test_scaled = StandardScaler().fit_transform(X_test)
```

# **Implementing Model & Evaluation**

**Logistic Regression**

```python
log_reg = LogisticRegression(max_iter= 1000) #default value in scikit 0.22 is 100
log_reg.fit(X_train_scaled, y_train.ravel())
```

**Evaluation**

```python
y_prob = log_reg.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_prob)
class_report = classification_report(y_test, y_prob)
conf_matrix = confusion_matrix(y_test, y_prob)

print(f"Accuracy Score: {accuracy*100} %")
print(f"\nClassification Report:")
print(class_report)
print(f"Confusion Matrix:")
print(conf_matrix)
```
